1.Above,I demostrated using a info() and shape functions with our lens_df data structure. * Describe what both of these functions are used for? * What information is returned from these functions? * Why is this information helpful? 

info() function gives list of non-null count and data type for our columns. It helps to identify what type of data each coulumn has like boolean, float64, int64, and object. Identifying data types can help us to know what coulumns need to be turn into numeric. It is important all data are numeric to perform further analysis. Similarly, knowing non-null count for each variable will help in data modication and filter. Less non-null counts in data tells us to either remove or add data. Moreover, this function also shows us the total memory usage. Similarly, shape function tells us the size of our dataframe. In fact, it gives the total number of rows and columns in our data. This way we can know the size of our data like number of variables/features with their values.

# Highest and lowest average rated movies?

```python
lens_df.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
ratings_df.groupby('movie_id').mean().sort_values(by='rating', ascending=False)
```

# Which movie has the most ratings in our dataset? 

```python
ratings_df.groupby('movie_id').sum().sort_values(by='rating', ascending=False)
```

```python
top_idx = ratings_df.groupby('movie_id').sum().sort_values(by='rating', ascending=False).index
movies_df.loc[top_idx]
```

Therefore, the most rated movies in our dataset is 'American Beauty(1999)' with the rating of 14800.

# List the 10 users who have rated the most movies?

```python
ratings_counts_df = lens_df.groupby(['user_id']).agg(user_count=('rating','count'), user_mean=('rating', 'mean')).reset_index()
ratings_counts_df
```

# Functions: * apply() * lambda() * aggregration (.agg) * groupby() * Conditional formatting in a dataframe

Using apply() function
Question: Make a separate column 'gender' and rename the 'male' as '99' and 'female' as '100'. Hint: use the apply () function.

```python
def return_gender(sex):
    if sex == 'male':
        return 99
    elif sex  == 'female':
        return 100
    else:
        return -999
```

```python
lens_df['gender'] = lens_df['sex'].apply(return_gender)
```

Usising grouby() and aggregation (.agg), find all possible measures?

```python
lens_df.groupby(['genres']).agg( {'rating':['sum', 'mean', 'median', 'min', 'max', 'std', 'var', 'mad', 'prod']} )
```

```python
lens_df.groupby(['genres']).agg ( {'rating' : 'describe'} ).round(2)
```


Using conditional formatting,  remove the decimals for sum, mean, median, and max.

```python
(lens_df.groupby(['genres'])
.agg( {'rating':['sum', 'mean', 'median', 'min', 'max']} )
.style.format('${0:,.0f}')
.highlight_max(color = 'green')
.highlight_min(color='red'))


```
